**Visualization of missing data points**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Missing data analysis
missing_data = corn_data.isnull().sum().sort_values(ascending=False)
missing_data = missing_data[missing_data > 0]  # Filter columns with missing data

# Plotting missing data
plt.figure(figsize=(10, 8))
sns.barplot(x=missing_data.values, y=missing_data.index)
plt.title('Missing Data Count in Each Column')
plt.xlabel('Number of Missing Values')
plt.ylabel('Columns')
plt.show()
```

**Outlier detection via Interquartile range**

```python
# Outlier detection using IQR for 'Yield'
Q1 = corn_data['Yield'].quantile(0.25)
Q3 = corn_data['Yield'].quantile(0.75)
IQR = Q3 - Q1

# Defining outliers as those beyond 1.5 times the IQR from the quartiles
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter outliers
outliers = corn_data[(corn_data['Yield'] < lower_bound) | (corn_data['Yield'] > upper_bound)]
outlier_free_corn_data = corn_data[(corn_data['Yield'] >= lower_bound) & (corn_data['Yield'] <= upper_bound)]
```

### Handling non-numeric enries: NULL

**Using the median values to impute the NULL cells**

```python
corn_data['Cation_Exchange'] = pd.to_numeric(corn_data['Cation_Exchange'], errors='coerce')

# Check if there are any missing values after the conversion
missing_cation_exchange = corn_data['Cation_Exchange'].isnull().sum()
print(f"Missing values in 'Cation_Exchange': {missing_cation_exchange}")

#imputing the NUll cells with the median value of the cation
median_cation_exchange = corn_data['Cation_Exchange'].median()
corn_data['Cation_Exchange'].fillna(median_cation_exchange, inplace=True)
```

```python
#This line of code confirms that there are no more missing values
missing_after_imputation = corn_data['Cation_Exchange'].isnull().sum()
print(f"Missing values in 'Cation_Exchange' after imputation: {missing_after_imputation}")
```

**Handling issue with the years that have 53 weeks**

```python
# This code shows tha 'W_prcp_53' through 'W_vp_53' are weekly measurements
# The cells with NaN represents weeks that do not exist.

# Create a mask for the years with 53 weeks
mask_53 = corn_data['year'].isin([2010, 2015, 2016, 2020, 2021])

# For years with only 52 weeks, set the 53rd week columns to zero
for column in corn_data.columns:
    if '53' in column:
        corn_data.loc[~mask_53, column] = 0
```

### Cluster analysis and visualization of the Corn Yield by County

**The current plot above for Time-Series Analysis Plot of Corn Yield visualization is crowded and hard to interpret because there are too many lines (one for each county).**

**I chose clustering of the counties based on similar yield patterns over the years to help reduce the complexity of the visualization**

**To begin, we use the Elbow Method in determining the number of clusters**

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


X = corn_data.T

#This line Determines the number of clusters using the Elbow method
sse = []
for k in range(1, 11):  
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)  # SSE for each k

# Plot the SSE values to find the elbow
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), sse, marker='o')
plt.title('Elbow Method to Determine Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')
plt.xticks(range(1, 11))
plt.show()
```

### K-means Clustering

```python

X = corn_data.pivot_table(index='loc_ID', columns='year', values='Yield').fillna(0)

kmeans = KMeans(n_clusters=3, random_state=0).fit(X)

#This line of code assigns the cluster labels to the dataframe
corn_data['cluster'] = corn_data['loc_ID'].map(lambda x: kmeans.labels_[x])

# Calculate the average yield for each cluster over the years
cluster_average = corn_data.groupby(['year', 'cluster'])['Yield'].mean().reset_index()

#Thi sline of code Plots the average yield over time for each cluster
plt.figure(figsize=(10, 6))
for cluster in sorted(cluster_average['cluster'].unique()):
    cluster_data = cluster_average[cluster_average['cluster'] == cluster]
    plt.plot(cluster_data['year'], cluster_data['Yield'], label=f'Cluster {cluster}')

plt.title('Average Corn Yield over Time by Cluster')
plt.xlabel('Year')
plt.ylabel('Average Yield (Bushels/acre)')
plt.legend()
plt.tight_layout() 
plt.show()
```

## Model Preparation

**Data Splitting**

```python
# Split data temporally
train_data = corn_data[corn_data['year'] <= 2018]
validation_data = corn_data[(corn_data['year'] >= 2019) & (corn_data['year'] <= 2020)]
test_data = corn_data[corn_data['year'] > 2020]
```

```python
# Check the shape of each subset
print("Training Set Size:", train_data.shape)
print("Validation Set Size:", validation_data.shape)
print("Testing Set Size:", test_data.shape)
```

```python
# Save the processed data
train_data.to_csv('train_data.csv', index=False)
validation_data.to_csv('validation_data.csv', index=False)
test_data.to_csv('test_data.csv', index=False)
```

```python
# Define your features and target variable names
features = corn_data.columns.difference(['year', 'Yield'])
target = 'Yield'

# Create the features and target variables for training
X_train = train_data[features]
y_train = train_data[target]

# Create the features and target variables for testing
X_test = test_data[features]
y_test = test_data[target]

# Create the features and target variables for validation
X_val = validation_data[features]
y_val = validation_data[target]
```

```python
# Shape Check
assert X_train.shape[0] == y_train.shape[0], "Mismatch in training rows"
assert X_test.shape[0] == y_test.shape[0], "Mismatch in test rows"
assert X_val.shape[0] == y_val.shape[0], "Mismatch in validation rows"
```

**Model Selection: Machine Learnign models for prediction**

**We start with the simple Random Forest and Gradient Boosting before tuning**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Fit the Random Forest model
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Predict on the test set
y_pred_rf = rf.predict(X_test)

# Calculate the RMSE for the Random Forest model
rmse_rf = mean_squared_error(y_test, y_pred_rf, squared=False)
print(f"Random Forest RMSE: {rmse_rf}")
```

```python
from sklearn.ensemble import GradientBoostingRegressor

# Fit the Gradient Boosting model
gb = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
gb.fit(X_train, y_train)

# Predict on the test set
y_pred_gb = gb.predict(X_test)

# Calculate the RMSE for the Gradient Boosting model
rmse_gb = mean_squared_error(y_test, y_pred_gb, squared=False)
print(f"Gradient Boosting RMSE: {rmse_gb}")
```

### Process of choosing the best Hyperparameter tuning straegy

**Setup Hyperparameter Grids**

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
import numpy as np

# Hyperparameter grid for Random Forest
rf_param_grid = {
    'n_estimators': np.arange(100, 1001, 100),
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt']
}
```

```python
# Randomized Search for Random Forest
rf_random_search = RandomizedSearchCV(
    estimator=RandomForestRegressor(random_state=42),
    param_distributions=rf_param_grid,
    n_iter=50, cv=5, verbose=2, random_state=42, n_jobs=-1, scoring='neg_mean_squared_error')
```

```python
# Fitting Random Forest
rf_random_search.fit(X_train, y_train)  # Ensure X_train, y_train are defined as your feature matrix and target variable
```

```python
# Best parameters and score for Random Forest
print("Best parameters for RF:", rf_random_search.best_params_)
print("Best RMSE for RF:", np.sqrt(-rf_random_search.best_score_))
```

```python
# Hyperparameter grid for Gradient Boosting
gb_param_grid = {
    'n_estimators': np.arange(100, 1001, 100),
    'learning_rate': [0.001, 0.01, 0.1, 0.2],
    'max_depth': [3, 5, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt']
}
```

```python
# Randomized Search for Gradient Boosting
gb_random_search = RandomizedSearchCV(
    estimator=GradientBoostingRegressor(random_state=42),
    param_distributions=gb_param_grid,
    n_iter=50, cv=5, verbose=2, random_state=42, n_jobs=-1, scoring='neg_mean_squared_error')
```

```python
# Fitting Gradient Boosting
gb_random_search.fit(X_train, y_train)
```

```python
# Best parameters and score for Gradient Boosting
print("Best parameters for GB:", gb_random_search.best_params_)
print("Best RMSE for GB:", np.sqrt(-gb_random_search.best_score_))
```
